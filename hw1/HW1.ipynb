To do this we will need two matrices.  Recall, first, the following formula for rigid body transformation
$$
^BP = \text{ } ^B_AR\text{ }^AP + \text{ } ^BO_A
$$

Where $^BP$ is the point coordinate in the target ($B$) coordinate system. $^AP$ is the point coordinate in the source ($A$) coordinate system. $^B_AR$ is the rotation matrix from $A$ to $B$, and $^BO_A$ is the origin of the coordinate system $A$ expressed in $B$ coordinates. 

The rotation and translation can be combined into a single 4 $\times$ 4 extrinsic parameter matrix, $P_e$, so that $^BP = P_e \text{ }^AP$ where $^AP$ and $^BP$ are in homogeneous coordinates.

Once transformed, the points can be photographed using the intrinsic camera matrix, $P_i$ which is a 3 $\times$ 4 matrix.

Once these are found, the image of a point, $^BP$, can be calculated as $P_i P_e \text{ }^AP$.

### 2.2: Image Calculation [12 points]

You are provided the following starter code as a guide to structure your code and plot the projected points. You are free to modify or use different functions for calculating the rigid transformations and the image projection but please use the plot_points() and main() functions provided.

Here is a list of the provided functions:

- to_homog(): converts points from Euclidean to homogenous coordinates
- from_homog(): converts points from homogenous coordinates back to Euclidean coordinates
- project_points(): the function that takes P_i, P_e, and the point coordinates and calculates the projection onto the camera
- camera_1(): function where you define you define your intrinsic and extrinsic parameters for each camera - there are also functions for camera 2,3,4
- plot_points(): plots the projected camera points
- main(): takes the camera coordinates


````python
import numpy as np
import matplotlib.pyplot as plt
import math

# convert points from Euclidian to homogeneous
def to_homog(points): 
    
    ## Your Code Here ##
    points_homog = np.copy(points)
    numCols = np.size(points, 1)
    homogRow = np.ones((1, numCols)) # row to append to create homog coord
    points_homog = np.vstack((points_homog, homogRow))
    
    return points_homog


# convert points from homogeneous to Euclidian
def from_homog(points_homog):
    
    ## Your Code Here ##
    points_eud = np.copy(points_homog).T
    points_eud = np.array([x[:-1]/x[-1] for x in points_eud])
    points_eud = points_eud.T
    
    return points_eud



# project 3D euclidian points to 2D Euclidian
def project_points(P_int, P_ext, pts):
    
    ## Your Code Here ##
    cameraMatrix = np.matmul(P_int, P_ext)
    homogMapping = np.matmul(cameraMatrix, to_homog(pts))
    projected_points = from_homog(homogMapping)
    
    return projected_points

# returns the rotation matrix about z then y axes
def getRotMatrix(yDeg, zDeg):
    
    sinY = math.sin(yDeg)
    cosY = math.cos(yDeg)
    
    sinZ = math.sin(zDeg)
    cosZ = math.cos(zDeg)
    
    firstRot = np.array([[cosZ, -sinZ, 0],
                         [sinZ,  cosZ, 0],
                         [0, 0, 1]])
    
    secondRot = np.array([[cosY, 0, sinY],
                          [0, 1, 0],
                          [-sinY, 0, cosY]])
    
    
    # general form of rotation matrix about z then y
    return np.matmul(secondRot, firstRot)

# returns the intrinsic matrix with focal length, focalLength
def getIntrinsic(focalLength):
    
    return np.array([[focalLength, 0, 0, 0],
                     [0, focalLength, 0, 0],
                     [0, 0, 1, 0]])

# returns the extrinsic matrix with rotation matrix, rotMatrix,
# and offset, offsetVec
def getExtrinsic(rotMatrix, offsetVec):
    
    extrinsic = np.copy(rotMatrix)
    extrinsic = np.hstack((extrinsic, offsetVec.T))
    extrinsic = np.vstack((extrinsic, np.array([[0,0,0,1]])))
    
    return extrinsic
````

````python
def rgb2gray(rgb):
    """ Convert rgb image to grayscale.
    """
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])

def getKey(pair):
    return pair[0]

# perform non maximum supression in window of size dim X dim
def NMS(eigs, dim):
    
    # matrix of suppressed values
    suppressed = eigs
    
    # list of values in the window
    windowVals = []
    
    # slide window across rows
    for row in range(0, np.size(suppressed, 0) - dim + 1):    
        # slide window across columns
        for col in range(0, np.size(suppressed, 1) - dim + 1):
            
            # put curr value in window as (value, (value row, value col))
            # into windowVals to sort on value later
            for winRowPos in range(row, row + dim):
                for winColPos in range(col, col + dim):
                    windowVals.append((suppressed[winRowPos][winColPos], (winRowPos, winColPos)))
                    suppressed[winRowPos][winColPos] = 0
            
            # get max value from the sorted windowVals
            maxVal = sorted(windowVals, key=getKey, reverse=True)[0]
            
            # get index of the max value in suppressed matrix
            valRow, valCol = maxVal[1]
            
            # set the index of the max value back to it's original value,
            # which was the max value in the window
            suppressed[valRow][valCol] = maxVal[0]
            windowVals = [] # reset array for sorting
    
    
    return suppressed

def corner_detect(image, nCorners, smoothSTD, windowSize, nmsSize):
    
    """Detect corners on a given image.

    Args:
        image: Given a grayscale image on which to detect corners.
        nCorners: Total number of corners to be extracted.
        smoothSTD: Standard deviation of the Gaussian smoothing kernel.
        windowSize: Window size for Gaussian smoothing kernel, and corner detector.

    Returns:
        Detected corners (in image coordinate) in a numpy array (n*2).

    """
    
    ### START CODE HERE###
    
    # clean image
    gauss_filter = gaussian2d(windowSize, smoothSTD)
    temp_image = np.copy(image)
    temp_image = convolve(temp_image, gauss_filter, mode='same')
    
    # filters to compute partial derivs for x and y
    xDerivFilter = np.array([[1, 0, -1]])
    yDerivFilter = np.array([[1, 0, -1]]).T
    
    # matrix of partial derivs for x and y. Shape of each same as image.
    xPartial = convolve(temp_image, xDerivFilter, mode='same')
    yPartial = convolve(temp_image, yDerivFilter, mode='same')
    
    # matrix of partial derivs squared for x and y. Shape of each same as image.
    xPartialSquare = np.square(xPartial)
    yPartialSquare = np.square(yPartial)
    
    # matrix of partialX * partialY in same index of partial matrix.
    # Shape same as image
    xyEleMult = xPartial * yPartial
    
    # window (filter) to sum everything in the window
    sumFilter = np.ones((windowSize, windowSize))
    
    # compute the sum for each in same window location
    xParSqSum = convolve(xPartialSquare, sumFilter, mode='same')
    yParSqSum = convolve(yPartialSquare, sumFilter, mode='same')
    xyPartSum = convolve(xyEleMult, sumFilter, mode='same')
    
    # hold the minimum eigval from each second moment matrix
    eigvals = []
    
#     print(image.shape)
#     print(xPartialSquare.shape)
#     print(yPartialSquare.shape)
#     print(xyPartSum.shape)
    
    # create the second moment matrix and append min eigen val to eigvals
    print('Calculating eigenvalues...')
    for (x, y, z) in zip(xParSqSum.flatten(), yParSqSum.flatten(), xyPartSum.flatten()):
        C = np.array([[x, z],
                      [z, y]])
        
        eig = np.linalg.eigvals(C)
        val = np.amin(eig)
        eigvals.append(val)
    
    # convert eigvals list to matrix with same dimensions as image
    eigvals = np.asarray(eigvals)
    eigvals = np.reshape(eigvals, image.shape)

    # perform non maximum supression on eigs matrix
    print('Performing non-maximum suppression...')
    eigvals = -NMS(eigvals, nmsSize)
    
    print('Getting {} corners...'.format(nCorners))
    sortedCorners = np.dstack(np.unravel_index(np.argsort(eigvals.ravel()), 
                                              (np.size(eigvals, 0), np.size(eigvals, 1))))[0]

    print('Detected {} corners.'.format(len(sortedCorners)))
    sortedCorners = sortedCorners[:,:: -1]
    
    #sortedCorners = np.asarray([np.flip(x) for x in sortedCorners])
    print('Done.')
    ### END CODE HERE ###
    
    return sortedCorners[0:nCorners]



# detect corners on the two provided images
# adjust your corner detection parameters here
nCorners = 100
smoothSTD = 2
windowSize = 11
nmsSize = 3

# read images and detect corners on images
imgs = []
corners = []
for i in range(2):
    img = imread('almond' + str(i) + '.jpg')
    imgs.append(rgb2gray(img))
    corners.append(corner_detect(imgs[-1], nCorners, smoothSTD, windowSize, nmsSize))

def show_corners_result(imgs, corners):
    fig = plt.figure(figsize=(16, 16))
    ax1 = fig.add_subplot(221)
    ax1.imshow(imgs[0], cmap='gray')
    ax1.scatter(corners[0][:, 0], corners[0][:, 1], s=35, edgecolors='r', facecolors='none')

    ax2 = fig.add_subplot(222)
    ax2.imshow(imgs[1], cmap='gray')
    ax2.scatter(corners[1][:, 0], corners[1][:, 1], s=35, edgecolors='r', facecolors='none')
    plt.show()

show_corners_result(imgs, corners)
````

### 4.2: Feature Matching using SIFT [10 points]

We have two overlapping images of a scene, and we would like to detect and match features (corners) across them. In order to do so, we will compute feature descriptors for the patches around the points and perform matching using a "sum of squared differences" metric.

First, you will want to implement a function which computes said metric. Given two vectors, `ssd` should return the sum of squared differences of their values:

$$ssd(v, w) = \sum_i (v_i - w_i)^2$$

````python
def ssd(v1, v2):
    """Compute SSD (sum of squared differences) for two NumPy arrays.
    This function will come in handy for descriptor matching later.
    """
    
    ### START CODE HERE ###
    diffSquare = np.square(v1 - v2)
    squareDistance = np.sum(diffSquare)
    
    return squareDistance
    ### END CODE HERE ###
````

You can check this using the ratio of the SSD distance with the closest match and the next-closest match. By Lowe's empirical findings, you should only accept the best match if the ratio between its distance and the second-best distance is ~0.75 or less.
